1. Retrieve a presidential speech from the Miller Center.

```python
import requests
from bs4 import BeautifulSoup

url = "https://millercenter.org/the-presidency/presidential-speeches/march-4-1861-first-inaugural-address"

html = requests.get(url).text

soup = BeautifulSoup(html, "lxml")

transcript = soup.find("div", {"class": "view-transcript"})

print(transcript.text)
```

2. Pre-process the text

```python
speech = transcript.text[11:]

speech = speech.replace("\n", " ")

import spacy
nlp = spacy.load('en_core_web_sm')

processed_speech = nlp(speech)

token_list = []

for t in processed_speech:
  if t.is_punct != True:
    token_list.append(t.text)

clean_speech = " ".join(t for t in token_list)

print(clean_speech)
```

3. Build a word cloud using the pre-processed speech.

```python
from wordcloud import WordCloud

import matplotlib.pyplot as plt

wc = WordCloud(min_font_size=16, max_words=100, background_color='white', width=600, height=600).generate(clean_speech)

plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()

wc.to_file("wc.png")
```

4. Interpret the results from the word cloud.

Based on the results of the word cloud, Abraham Lincoln's first inaugural address revolved much around words such as "Constitution", "will", "law", and "government". His speech seemingly had a great deal of emphasis surrounding the Constitution and future implementation of law. In addition, words such as "people", "slave", "right" and "provision" leads one to believe these changes to law and government seemingly would benefit slaves and the American people as a whole. This is very consistent with the time and with what Lincoln ultimately accomplished for America during his presidency.
