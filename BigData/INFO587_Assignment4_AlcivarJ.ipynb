**QUESTION 2.2:** 
* Write code to remove punctuation and numbers from each tweet located in the textRDD we just created above

Explanation
* Real world files are complicated. Some of the issues we have to address are:
* Words should be counted independent of their capitialization (e.g., Spark and spark should be counted as the same word).
* All punctuations should be removed.
* Any leading or trailing spaces on a line should be removed.
* Define the function removePunctuation that converts all text to lower case, removes any punctuation, and removes leading and trailing spaces as well as numbers. Use the Python re module to remove any text that is not a letter, number, or space. Reading help(re.sub) might be useful. If you are unfamiliar with regular expressions, you may want to review tutorials on Google.

```
# Your code goes in this cell - program the function to remove punctuation and numbers from texts

import string
import re

def removePunctuation(text):
    """
    YOUR CODE GOES IN HERE
    Removes punctuation, changes to lower case, and strips leading and trailing spaces.

    Note:
        Only spaces, letters, and numbers should be retained.  Other characters should should be
        eliminated (e.g. it's becomes its).  Leading and trailing spaces should be removed after
        punctuation is removed.

    Args:
        text (str): A string.

    Returns:
        str: The cleaned up string.
    """
    text = re.sub(r'[^\w\d\s]+', '', text)
    text = re.sub(r'[0-9]+', '', text).strip()
    text = text.lower()
    
    return text
  
#################
# do not change anything in this cell. This will test your code
textRDD = textRDD.map(lambda x: removePunctuation(x)) # this calls the function and removes the punctuation keep this as is - do not change it do not delete it.
textRDD.take(2) # keep this as is - do not change it do not delete it
```
