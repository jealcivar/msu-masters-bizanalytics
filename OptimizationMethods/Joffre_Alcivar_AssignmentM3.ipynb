```python
from pyomo.environ import *
```

# Question 1

```python
# Create model
model1 = ConcreteModel()

# Declare variables
model1.x1 = Var(domain=NonNegativeReals)
model1.x2 = Var(domain=NonNegativeReals)

# Declare Objective function
model1.Z = Objective(expr = 2*model1.x1 + 1*model1.x2, sense=maximize)

# Declare constraints
model1.const1 = Constraint(expr = 0*model1.x1 + 1*model1.x2 <= 10)
model1.const2 = Constraint(expr = 2*model1.x1 + 5*model1.x2 <= 60)
model1.const3 = Constraint(expr = 1*model1.x1 + 1*model1.x2 <= 18)
model1.const4 = Constraint(expr = 3*model1.x1 + 1*model1.x2 <= 44)

# Solve
results = SolverFactory('glpk').solve(model1)
results.write()
model1.pprint()

# Display solutions
print('\nMax Z = ', model1.Z())
print('\nx1 = ', model1.x1())
print('x2 = ', model1.x2())
```

# Question 2

```python
# Create model
model2 = ConcreteModel()

# Declare variables
model2.x1 = Var(domain=NonNegativeReals)
model2.x2 = Var(domain=NonNegativeReals)
model2.x3 = Var(domain=NonNegativeReals)
model2.x4 = Var(domain=NonNegativeReals)
model2.x5 = Var(domain=NonNegativeReals)

# Declare Objective function
model2.cost = Objective(expr = 22*model2.x1 + 20*model2.x2 + 25*model2.x3 + 24*model2.x4 + 27*model2.x5, 
                       sense=minimize)

# Declare constraints
model2.tin = Constraint(expr = 60*model2.x1 + 25*model2.x2 + 45*model2.x3 + 20*model2.x4 + 50*model2.x5 == 40)
model2.zinc = Constraint(expr = 10*model2.x1 + 15*model2.x2 + 45*model2.x3 + 50*model2.x4 + 40*model2.x5 == 35)
model2.lead = Constraint(expr = 30*model2.x1 + 60*model2.x2 + 10*model2.x3 + 30*model2.x4 + 10*model2.x5 == 25)

# Solve
results = SolverFactory('glpk').solve(model2)
results.write()
model2.pprint()

# Display solutions
print('\nCost = ', model2.cost())
print('\nx1 = ', model2.x1())
print('x2 = ', model2.x2())
print('x3 = ', model2.x3())
print('x4 = ', model2.x4())
print('x5 = ', model2.x5())
```

# Question 3

```python
# Create model
model3 = ConcreteModel()

# Declare variables
model3.x1 = Var(domain=NonNegativeReals)
model3.x2 = Var(domain=NonNegativeReals)

# Declare Objective function
model3.profit = Objective(expr = 32*model3.x1 + 24*model3.x2, sense=maximize)

# Declare constraints
model3.nylon = Constraint(expr = 3*model3.x1 + 2*model3.x2 <= 5000)
model3.salesx1 = Constraint(expr = 1*model3.x1 + 0*model3.x2 <= 1000)
model3.salesx2 = Constraint(expr = 0*model3.x1 + 1*model3.x2 <= 1200)
model3.labor = Constraint(expr = 45*model3.x1 + 40*model3.x2 <= 84000)

# Solve
results = SolverFactory('glpk').solve(model3)
results.write()
model3.pprint()

# Display solutions
print('\nProfit = ', model3.profit())
print('\nx1 = ', model3.x1())
print('x2 = ', model3.x2())
```

# Question 4

```python
# Create model
model4 = ConcreteModel()

# Declare variables
model4.x1 = Var(domain=NonNegativeReals)
model4.x2 = Var(domain=NonNegativeReals)

# Declare Objective function
model4.cost = Objective(expr = 1.20*model4.x1 + 1.70*model4.x2, sense=minimize)

# Declare constraints
model4.octane = Constraint(expr = 88*model4.x1/88 + 66*model4.x2/66 >= 78)
model4.fueltotal = Constraint(expr = 1*model4.x1 + 1*model4.x2 >= 3300)
model4.fuela = Constraint(expr = 1*model4.x1 + 0*model4.x2 <= 2200)
model4.fuelb = Constraint(expr = 0*model4.x1 + 1*model4.x2 <= 3600)
model4.storage = Constraint(expr = 1*model4.x1 + 1*model4.x2 <= 4300)

# Solve
results = SolverFactory('glpk').solve(model4)
results.write()
model4.pprint()

# Display solutions
print('\nCost = ', model4.cost())
print('\nx1 = ', model4.x1())
print('x2 = ', model4.x2())
```
